```python
# prompt: prompt: 根据刚刚给的catagory以及对应的product，以及价格，生成一个1500行3列的数组，每10行（比如1到40行）为这40个catagory的随机选取10个并打乱顺序排列排列，第二列为对应的product，第三列为对于价格的一个随机变动，这个变动可以为原价乘以[0%，20%，40%，60%，80%，100%，120%，140%，160%，180%，200%]，即为原价格乘以对应比例,最后告诉我行列数目

import pandas as pd
import numpy as np
import random

# Assuming 'catagory.csv' has columns 'category', 'product', and 'price'
# and is already loaded into the 'catagory' DataFrame as shown in the previous code.

# Create an empty list to store the data
data = []

# Define possible price multipliers
price_multipliers = [0, 0.2, 0.4, 0.6, 0.8, 1, 1.2, 1.4, 1.6, 1.8, 2]

for i in range(200):  # Generate data for 150 sets of 10 rows
    # Randomly select 10 unique categories
    selected_categories = random.sample(range(len(catagory)), 10)
    for idx in selected_categories:
        category = catagory['Category'].iloc[idx]
        product = catagory['Product'].iloc[idx]
        original_price = catagory['Price($)'].iloc[idx]
        price_multiplier = random.choice(price_multipliers)
        new_price = original_price * price_multiplier
        data.append([category, product, new_price])

# Create a Pandas DataFrame from the collected data
result_df = pd.DataFrame(data, columns=['Category', 'Product', 'Price'])

# Print the shape of the resulting DataFrame
print("Shape of the DataFrame:", result_df.shape)

# Display the first few rows of the DataFrame
print(result_df.head())

# Save to csv
result_df.to_csv('output.csv', index=False)
```

```python
def ask_gpt(prompts):
    responses = []

    for prompt_index, prompt in enumerate(prompts[:]):
        print(f"Sending prompt {prompt_index + 1}...")  # Display which prompt is being processed

        for i in range(50):  # Send each prompt 50 times
            response = client.chat.completions.create(
                model="gpt-4o-mini",  # Use the GPT-4 mini model
                messages=[
                    {"role": "system", "content": "You, AI, are a customer. Your task is to fill in the blanks_______, Return complete information in comma-separated values, without any extra text."},
                    {"role": "user", "content": prompt}
                ],
                temperature=1,  # Set temperature to 1
                max_tokens=5000  # Adjust max_tokens if needed
            )

            # Correct way to access the content from the response
            response_content  = response.choices[0].message.content

            # Store all responses in a list
            responses.append(response_content)

    return responses

# Example usage
responses = ask_gpt(prompts)
```
